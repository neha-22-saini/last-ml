Question 1) What is Dimensionality Reduction? Why is it important in machine
learning?

Ans.
# **Why Dimensionality Reduction is Important in Machine Learning?**

### **1. Reduces Overfitting**

* When too many features exist, models may “memorize” noise instead of learning patterns.
* Reducing dimensions removes irrelevant attributes → improves generalization.

### **2. Improves Model Performance**

* Less features → faster training
* Models become simpler and more efficient
* Reduces memory and computation cost
  (Especially important for algorithms like KNN, SVM, clustering, etc.)

### **3. Helps Visualization**

* Human beings can visualize only **2D or 3D**.
* Dimensionality reduction techniques (PCA, t-SNE, UMAP) reduce data to 2 or 3 dimensions → helps in understanding patterns, clusters, and anomalies.

### **4. Removes Noise and Multicollinearity**

* Many features contain redundant or highly correlated information.
* Removing these improves model stability and accuracy.

### **5. Reduces Storage and Improves Efficiency**

* Storing high-dimensional datasets requires more space.
* Compressed representations consume less storage and speed up processing.

### **6. Helps Avoid the “Curse of Dimensionality”**

* Distance-based algorithms (KNN, clustering) perform poorly in high dimensions.
* Dimensionality reduction reduces sparsity and improves algorithm effectiveness.



Question 2: Name and briefly describe three common dimensionality reduction
techniques..

Ans
1. Principal Component Analysis (PCA)

PCA is a linear dimensionality reduction method.

It transforms the original features into a new set of uncorrelated features called principal components.

These components capture the maximum variance in the data.

Most important components are kept, and the rest are discarded.

2. Linear Discriminant Analysis (LDA)

LDA is a supervised technique (uses class labels).

It aims to find a lower-dimensional space where class separation is maximized.

It projects data in such a way that the distance between classes increases and within-class variance decreases.

Often used in classification tasks like face recognition.

3. t-Distributed Stochastic Neighbor Embedding (t-SNE)

t-SNE is a non-linear, probabilistic dimensionality reduction technique.

Especially useful for visualizing high-dimensional data in 2D or 3D.

It preserves local structure (similar points stay close).

Commonly used in deep learning, genomics, and clustering visualization.

| Technique | Type                | Key Idea                        | Used For                   |
| --------- | ------------------- | ------------------------------- | -------------------------- |
| **PCA**   | Linear              | Maximize variance               | Compression, noise removal |
| **LDA**   | Linear + Supervised | Maximize class separability     | Classification tasks       |
| **t-SNE** | Non-linear          | Preserve neighborhood structure | 2D/3D visualization        |


Question 3: What is clustering in unsupervised learning? Mention three popular
clustering algorithms.

Ans.
    Definition: Clustering

Clustering is an unsupervised learning technique used to group similar data points into clusters based on their similarity, distance, or patterns in the data.
In clustering, the algorithm does not use labeled outputs—it simply discovers natural structures or groups present in the dataset.

Goal:
To ensure that

Data points within the same cluster are similar, and

Data points in different clusters are dissimilar.

Three Popular Clustering Algorithms:
1. K-Means Clustering

Partitions data into K clusters.

Uses centroids to represent each cluster.

Iteratively assigns points to the nearest centroid and recomputes centroids.

2. Hierarchical Clustering

Builds a tree-like structure (dendrogram).

Two types:

Agglomerative (bottom-up)

Divisive (top-down)

No need to specify number of clusters initially.

3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

Groups points based on density.

Can find arbitrary-shaped clusters.

Identifies noise/outliers effectively.

Works well with uneven cluster sizes.

Question 5: List and briefly describe three types of anomaly detection techniques.


Ans.
    1. Statistical (Distribution-Based) Methods

Identify anomalies based on probability distributions.

Data points that fall far from the mean, or outside expected ranges (e.g., ±3 standard deviations) are considered anomalies.

Examples: Z-score, Gaussian models.

2. Distance-Based Methods

Assume that normal points lie in dense regions, while anomalies are far from other data points.

Points with large distance from their neighbors are flagged as anomalies.

Examples: K-Nearest Neighbors (KNN), Mahalanobis distance.

3. Density-Based Methods

Use the idea that anomalies occur in sparse regions where data density is low.

Identify points that have significantly lower density compared to neighbors.

Examples: DBSCAN, Local Outlier Factor (LOF).

Question 6: What is time series analysis? Mention two key components of time series data.

Ans.
Time Series Analysis (Definition)

Time series analysis is the process of examining and modeling data collected over time to identify patterns such as trends, cycles, and seasonal variations.
It helps in forecasting future values, understanding past behavior, and detecting changes in temporal data.

Examples: stock market prices, temperature readings, daily sales, electricity usage.

Two Key Components of Time Series Data
1. Trend

A long-term increase or decrease in the data over time.

Shows the overall direction of the series (upward, downward, or stable).

2. Seasonality

Regular, repeating patterns that occur at fixed intervals (daily, monthly, yearly).

Example: higher retail sales during festivals every year.

Question 7: Describe the difference between seasonality and cyclic behavior in time
series.



* Seasonality refers to **regular, repeating patterns** in time series data that occur at **fixed and known intervals**.
* These intervals can be **daily, weekly, monthly, quarterly, or yearly**.
* The pattern is **predictable** and caused by external factors such as weather, holidays, or human behavior.

**Example:**

* Higher ice-cream sales every summer.
* Increased shopping during festivals every year.

---

### **Cyclic Behavior**

* Cyclic behavior refers to patterns that **rise and fall irregularly over long periods**, typically over **more than one year**.
* Cycles do **not have a fixed period** and are influenced by **economic or business conditions**.
* The pattern is **not strictly predictable**; the length and intensity of cycles can vary.

**Example:**

* Economic boom and recession cycles.
* Long-term business growth and decline phases.

---

# **Key Differences (Summary)**

| Feature        | Seasonality                     | Cyclic Behavior               |
| -------------- | ------------------------------- | ----------------------------- |
| Frequency      | Fixed, regular intervals        | Irregular, variable intervals |
| Predictability | High, predictable               | Low, not predictable exactly  |
| Time Span      | Short-term (days/months/year)   | Long-term (years)             |
| Cause          | Weather, holidays, human habits | Economic/market forces        |

Q8: Write Python code to perform K-means clustering on a sample dataset.

```python
# K-means Clustering Example in Python

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 1: Create a sample dataset
X, y = make_blobs(n_samples=300, centers=3, random_state=42)

# Step 2: Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Step 3: Get cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Step 4: Plot the clustered data
plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(centroids[:,0], centroids[:,1], marker='X', s=200)
plt.title("K-Means Clustering")
plt.show()
```

Question 9: What is inheritance in OOP? Provide a simple example in Python.


Ans.
    Definition: Inheritance

Inheritance is an OOP (Object-Oriented Programming) concept where a child class (subclass) acquires the properties and methods of a parent class (superclass).
It promotes code reuse, extensibility, and organizational structure in programs.

```python
# Parent class
class Animal:
    def sound(self):
        return "Animals make different sounds"

# Child class inheriting from Animal
class Dog(Animal):
    def sound(self):
        return "Dog barks"

# Using the classes
a = Animal()
d = Dog()

print(a.sound())   # Output: Animals make different sounds
print(d.sound())   # Output: Dog barks
```

Question 10: How can time series analysis be used for anomaly detection?

Ans.
    
Time series analysis can be used for anomaly detection by examining **patterns over time**—such as trend, seasonality, and normal fluctuations—to identify data points that **deviate significantly** from expected behavior.

### **How It Works:**

### **1. Identify Expected Patterns**

Time series analysis models the **normal behavior** of data using:

* Trend
* Seasonality
* Cycles
* Autocorrelation

Once the normal pattern is known, we can compare new data against it.

---

### **2. Detect Deviations from the Model**

If a data point **does not follow the expected pattern**, it is marked as an anomaly.

Examples:

* Sudden spike in server traffic
* Unexpected drop in sales
* Unusual temperature reading

---

### **3. Use Time-Series Models for Anomaly Detection**

#### **a) ARIMA / SARIMA Models**

* Predict future values.
* If actual value deviates from the predicted value beyond a threshold → anomaly.

#### **b) Moving Average (MA) & Exponential Smoothing**

* Detect abnormal spikes or drops by comparing with smoothed values.

#### **c) Seasonal Decomposition (STL)**

* Breaks time series into:
  **Trend + Seasonality + Residual**
* Anomalies appear in the **residual** component.

#### **d) Machine Learning Models**

* LSTM (Long Short-Term Memory networks)
* Autoencoders
  Can learn complex temporal patterns and detect anomalies when reconstruction error is high.

---

# **Why Time Series Analysis Helps in Anomaly Detection?**

* Captures **temporal dependencies**
* Distinguishes true anomalies from **seasonal changes**
* Reduces **false alarms**
* Useful for forecasting normal vs abnormal behavior

---

# **Examples**

* Detecting a sudden fraud transaction
* Identifying abnormal heart rate in health monitoring
* Detecting equipment failure using sensor data
* Spotting unusual website traffic spikes
